```julia
m_x_0 = 0.0
v_x_0 = 0.5
v_x_0 = 1.0
v_v_0 = 0.1


m_x = Vector{Float64}(undef, n_samples)
v_x = Vector{Float64}(undef, n_samples)
m_v = Vector{Float64}(undef, n_samples)
v_v = Vector{Float64}(undef, n_samples)
m_x_t_min = m_x_0
v_x_t_min = v_x_0
m_v_t_min = 1
v_v_t_min = 0.1

for t = 1:n_samples
    # Prepare data and prior statistics
    data = Dict(:z_t       => z[t],
                :m_v_t_min => m_v_t_min,
                :v_v_t_min => v_v_t_min,
                :m_x_t_min => m_x_t_min,
                :v_x_t_min => v_x_t_min)
    # Execute algorithm
    marginals_v = step_v_fwd!(data)
    marginals_x = step_x_fwd!(data)
    
    m_x_t_min = mean(marginals_x[:x_t])
    v_x_t_min = var(marginals_x[:x_t])
    m_v_t_min = mean(marginals_v[:v_t])
    v_v_t_min = var(marginals_v[:v_t])
    m_x[t] = m_x_t_min
    v_x[t] = v_x_t_min
    m_v[t] = m_v_t_min
    v_v[t] = v_v_t_min
end
```
